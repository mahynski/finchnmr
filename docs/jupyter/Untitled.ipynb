```python
import finchnmr
from finchnmr import library, substance, model
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os

substances = []
failed = []
head = '../../../BMRB_Database/'
for metabolite_ in os.listdir(head):
    print(metabolite_)
    pathname_ = os.path.join(
        os.path.abspath(
            os.path.join(
                head, 
                metabolite_
            )
        ), '1/pdata/1'
    )
    try:
        s = finchnmr.substance.Substance(pathname_, name=metabolite_, warning='default')
    except Exception as e:
        print(e)
        failed.append(metabolite_)
    else:
        substances.append(s)
```

# Analyze various samples

```python
head = '../../../Fish Liver HSQC spectra/'
targets = []
for sample_ in os.listdir(head):
    pathname_ = os.path.join(
        os.path.abspath(
            os.path.join(
                head, 
                sample_
            )
        ), 'pdata/1'
    )
    targets.append(
        substance.Substance(pathname_)
    )
```

```python
optimized_models, analyses = finchnmr.model.optimize_models(
    targets=targets,
    nmr_library=lib,
    nmr_model=finchnmr.model.LASSO,
    param_grid={'alpha': np.logspace(-16, 0, 5)},
    model_kw={'max_iter':1000, 'selection':'cyclic', 'random_state':42, 'tol':0.0001}
)
```

```python
model.plot_stacked_importances(optimized_models, figsize=(2,10))
plt.tight_layout()
```
